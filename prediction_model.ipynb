```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import numpy as np
```

```python

df = pd.read_csv("C:/Users/kanak/OneDrive/Desktop/dataset9000.csv")


X = df.drop("Role", axis=1)
y = df["Role"]

skill_levels = {
    'Not Interested': 0,
    'Poor': 1,
    'Beginner': 2,
    'Average': 3,
    'Intermediate': 4,
    'Excellent': 5,
    'Professional': 6
}
X_encoded = X.replace(skill_levels)

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
def predict_top_3_roles(user_input: dict):
    """
    Predicts top 3 career roles for given user skill input.
    `user_input` should be a dict with keys matching column names and skill levels as values.
    Example:
    {
        "Database Fundamentals": "Intermediate",
        "AI ML": "Excellent",
        ...
    }
    """
    user_df = pd.DataFrame([user_input])
    user_df_encoded = user_df.replace(skill_levels)

    probs = rf_model.predict_proba(user_df_encoded)[0]

    top_3_indices = np.argsort(probs)[-3:][::-1]
    top_3_roles = label_encoder.inverse_transform(top_3_indices)
    top_3_probs = probs[top_3_indices]

    return list(zip(top_3_roles, top_3_probs))
```

```python
# Example user input
new_user_input = {
    "Database Fundamentals": "Intermediate",
    "Computer Architecture": "Average",
    "Distributed Computing Systems": "Beginner",
    "Cyber Security": "Excellent",
    "Networking": "Poor",
    "Software Development": "Excellent",
    "Programming Skills": "Excellent",
    "Project Management": "Average",
    "Computer Forensics Fundamentals": "Beginner",
    "Technical Communication": "Intermediate",
    "AI ML": "Excellent",
    "Software Engineering": "Excellent",
    "Business Analysis": "Average",
    "Communication skills": "Excellent",
    "Data Science": "Excellent",
    "Troubleshooting skills": "Intermediate",
    "Graphics Designing": "Poor"
}

# Get predictions
top_3 = predict_top_3_roles(new_user_input)
for i, (role, prob) in enumerate(top_3, 1):
    print(f"Top {i} Role: {role} (Confidence: {prob:.2f})")

    
```
